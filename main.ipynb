```python
import pandas as pd
```

```python
df = pd.read_csv('ids.csv')
```

```python
import requests
import time
import os

download_folder = "nist_files/"
def download_files_from_urls():
    for i, nist_id in enumerate(df['NIST ID']):
        url_0 = f"https://webbook.nist.gov/cgi/cbook.cgi?JCAMP={nist_id}&Index=0&Type=IR"
        url_1 = f"https://webbook.nist.gov/cgi/cbook.cgi?JCAMP={nist_id}&Index=1&Type=IR"
        
        filename = os.path.join(download_folder, f"file_{i+1}.jdx")

        try:
            # Try downloading with url_1 first
            response = requests.get(url_1)
            response.raise_for_status()
            
            # If successful, save the file
            with open(filename, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded file {i+1} using url_1")
        except requests.exceptions.RequestException:
            try:
                # If url_1 fails, try url_0
                response = requests.get(url_0)
                response.raise_for_status()
                
                # If successful, save the file
                with open(filename, 'wb') as file:
                    file.write(response.content)
                print(f"Downloaded file {i+1} using url_0")
            except requests.exceptions.RequestException as e:
                # If both urls fail, print an error message
                print(f"Error downloading ID: {nist_id}")
        
        # Add a delay to avoid overwhelming the server
        time.sleep(5)
```

```python
download_files_from_urls()
```
